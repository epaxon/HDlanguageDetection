## 2. Alternative encodings

The use of representation as high-dimensional (N) binary vectors ($[-1, +1] ^ N$ -- often used as unitary binary vectors $[-1/N,+1/N]^N$), super-position as addition ($+$), similarity as dot product ($\cdot$) binding as element-wise multiplication ($*$), and permutation as `roll` ($r$) is one complete computational algebra of many that can be used for hypercomputing. But there are other possible frameworks, and I would like to get a thorough handle on all hypercomputing theories.

### 2.1 Holographic Reduced Representations

An alternative framework, developed by Plate, uses hypervectors chosen at random from a Normal distribution -- "normal-vectors" ($[\mathscr{N}(0,1/N)]^N$). This is very similar (basically the same) as the framework used by Eliasmith, which has basic elements chosen from the unit hypersphere -- "sphere-vectors" ($[\mathscr{S(1)}]^N$). The super-position operation remains addition, similarity remains dot-product (or very closely cosine similarity), but the binding operation is transformed into "circular convolution" ($\oplus$).

Here, I will attempt to perform the same memory experiments and explore whether the alternative encodings have any advantages. But more importantly, we will get a sense of hypercomputing more generally, because each of these encodings are just slightly different ways of doing the same thing, namely computation.

Storing the statistics in HRR hypervectors is very simple and is pretty much exactly the same as the hypercomputing framework. This is because we have not needed to use the binding operation thus far. The $+$ and similarity operators ($\cdot$) are the same in both frameworks!

However, when we start using the binding operation, $\oplus$, there are two key differences. First, we have to implement the binding operation funciton, which requires extra horsepower compared to $r$ and $*$. Second, in the hypersphere encoding, the vectors are no longer their self inverses. With encoding in which all vectors are $[-1, +1]$, each hypervector representation is its self inverse (i.e. $A*A=I$, where $I$ is the all +1 vector.) 

We can see this is true:

```python
k_vecs = 2 * (np.random.randn(len(random_idx.alphabet), N) < 0) - 1

k_vecs * k_vecs # all 1s
```

```python
sph_vecs = hs.sample(len(random_idx.alphabet), N)

sph_vecs * sph_vecs
```

But in this framework, the vectors are not their self-inverses, and element-wise multiplication is not the appropriate binding operation. Instead, we must encode and decode using circular convolution and the appropriate circular convolution inverse. 


```python
def cconv(a, b):
    return real(ifft(fft(a) * fft(b)))
    
```

```python
def ccinv(a):
    return real(ifft(np.conj(fft(a))))
```

Now we can bind and unbind vectors like we would for using element-wise multiplication. 

```python
s = cconv(sph_vecs[0,:], sph_vecs[1,:])

s2 = cconv(s, ccinv(sph_vecs[0,:]))

print np.dot(s2, sph_vecs[1,:])
```

The circular convolution's identity is a vector that is $[1, 0, 0, ...]$

```python
np.round(np.real(cconv(sph_vecs[0,:], ccinv(sph_vecs[0,:]))))
```

Wonderful! Now, we have all of the operations needed for plate's HRR implemented.

Plate's framework, however, has a fundamental limitation compared to the Kanerva framework. Plate hypervectors lack a permutation operation. This causes issues when trying to store the statistics of bigrams and trigrams, which need permutation and binding to create hypervector representations of ordered pairs or triplets of letters.

There were several possible implementations of roll that I have attempted, but each has failed. At first, I thought I could bind an extra `roll_vector` ($rv$) to the letters, but this does not work because $\oplus$ is commutative and associative:

$$
    rv \oplus a \oplus b = a \oplus rv \oplus b = rv \oplus b \oplus a
$$

and so this does not create a distinguishing hypervector that differentiates $ab$, from $ba$. 

Another option is to use the same permutation operation ($r$, `np.roll`). But again this is problematic for the same reason -- the permutation operation is also commutative and associative over *circular convolution*. This means that:

$$
    r(a) \oplus b = a \oplus r(b)
$$

This is unlike Kanerva's framework, where permutation is not associative across *multiplication*:

$$
    r(a) * b \ne a * r(b)
$$

This actually is a severe limitation, and there doesn't seem to be a good way of creating proper sequences in Plate's framework. Plate discusses several ways of storing sequences, yet each would have problems with this particular task.

Knowing this will not work, lets just see what happens.

**Figure 2.7. Sequences are a challenge in Plate's framework:** Top. An HRR hypervector was used to store the statistics of bigrams utilizing a unique roll vector to bind and create an ordered bigram. However, because circular is commutative, this operation results in the same hypervector representation for the sequence $ab$ and $ba$. Thus the stats for $ab$ and $ba$ are accumulated together, and this can be seen in the plots by the fact that each bigram and its mirror is on the same horizontal level (i.e. they have the same value in the hypervector). Bottom. The same was done but using the permutation operation ($r$), which also fails in the same fashion because permutation is also commutative over circular convolution.

So we can see in Fig. XX above that the bigram pairs are not distinguished in the hypervector (i.e. the values for $\_e$ and $e\_$ are the same). You can see that every bigram's reflection is on the same horizontal line -- meaning that the reflection has two different values in the real count, but the same value in the hypervector. This is all expected since neither of these methods actually distinguishes $ab$ from $ba$. 

Plate does discuss several options for creating sequences using the circular convolution and addition. However, each of these will be problematic in our attempts to collect stats about n-grams. In Plate 2003, he discusses several proposals for sequences from different papers in the literature. The first proposed by Murdock (1983, 1987) and Lewandowsky and Murdock (1989) uses a chaining method -- to store $abc$:

$$
    v = \alpha_1 a + \beta_1 a \oplus b + \alpha_2 b + \beta_2 b \oplus c + \alpha_3 c
$$

where $\alpha_i$ and $\beta_i$ are weighting parameters, with $\alpha_i > \alpha_{i+1}$. This type of representation has certain properties that make it nice and potentially useful for storing a sequence, but it is fundamentally not the same as the permutation and binding operation used in Kanerva's framework. There are other similar ways he describes, but these all are based on super-positions of bound elements.

Plate also proposes something called "trajectory association", which binds each element of a sequence to another hypervector that uniquely represents each position:

$$
    v = p_1 \oplus a + p_2 \oplus b + p_3 \oplus c
$$

which can use an arbitrary vector $p$, which can be then raised to the $i$ th convolutional power (e.g. $p_i = (p)^i$ to make each position vector. But again this representation has issues with storing the statistics of n-grams because the ngrams with letters at the same position will interfere, for instance look at a vector that sees 5 $the$'s and 2 $why$'s being stored together:

$$
    v = 5*(p_1 \oplus t + p_2 \oplus h + p_3 \oplus e) + 2*(p_1 \oplus w + p_2 \oplus h + p_3 \oplus y)
$$

which has the h values interfering with each other because $v$ can be rewritten:

$$
    v = 5 * p_1 \oplus t + 5 * p_3 \oplus e + 2 * p_1 \oplus w + 2 * p_3 \oplus y + \mathbf{7 * p_2 \oplus h}
$$

so this vector will just be weird and it won't be possible to decode the values of particular n-grams. 

Generally, to build up new representations of pieces that depend on the order of the pieces, one needs a non-commutative binding operation. 

### 2.2 Plate and Kanerva

What I am going to show next is that all of the operators still apply in both frameworks and that both frameworks are more similar than I originally thought. We can use element-wise multiplication ($*$) for binding, and roll ($r$) for permutation just as well using sphere-vectors as with unitary binary vectors ($[+1/N, -1/N]^N$). However, the essential problem with using $*$ in Plate's framework (and likely the reason circular convolution was developed) is that the inverse ($a^{-1}$) is *unstable*. Because the hypervectors in Plate's framework can have 0 values (or extremely small values), the inverses have extremely *large* values, that can easily blow up the hypervector with a small amount of noise. This is not the case in the Kanerva framework, because the vector are unitary, and hence their inverses are also unitary (and thus stable).

So, as long as we don't really need to use the inverse, we can use the standard Kanerva operations ($+, *, r$) on the Plate hypervectors (we also have to rescale for each multiplication by $\sqrt{N}$, and/or renormalize the hypervectors after each multiply). When we do bigrams or trigrams, the algorithm we developed used an extra hypervector to keep track of the current n-gram. An alternative would be to just keep track of each letter, and then generate the hypervector each time. The latter method, in my opinion, is a bit cheating in terms of making a general hypervector computer, because we need memory for non-hypervectors. But the former method does require utilizing the inverse operation with Plate's representation, and this will typically lead to the hypervector blowing up. But careful re-normalization can make it work.

It works exactly the same, so long as normalization of the inverses is allowed. But lets dig a little deeper and see how the inverse can mess things up. 

Using the inverse is ok for 3-grams, so long as normalization is available. The other thing that helps is that in this situation the inverse *always works* -- in that the `ngram_vector` always has the original vector present. 

The problems with the inverse of multiplication in HRR framework is very apparent when a hypervector is dereferenced by an inverse that was not stored. If the $a$ is present in a hypervector, then multiplying by the inverse of $a$ is usually kosher. However, if $a$ is not present, then the inverse can lead to big problems due to the unstable inverse, unless normalization is available. 

The main aspect needed to get multiply to work with Plate is to renormalize. If the vectors are not renormalized, and only the expected normalization is used, then there can be dramatic consequences. With sphere-vectors, multiplication results in a smaller vector, and so they need to be renormalized by rescaling by $\sqrt N$. Multiplying by an inverse vector needs to be rescaled by $1/ \sqrt N $. But the inverse rescaling can be extremely noisy, and for it to work at all full renormalization is needed.

To illustrate this I will make two bound hypervectors, one that stores the bigram $ab = r(a)*b$, and the other that stores $ba=r(b)*a$. Then we will dereference these hypervectors with the inverse of $b$ in the first position, $r(b)^{-1}$. This should lead to null similarities in the dereferenced $ab$ hypervector, and a high dot-product with $a$ in the dereferenced $ba$ hypervector.


```python
ab = np.roll(letter_vectors[0,:], 1) * letter_vectors[1,:] * N ** 0.5
ba = np.roll(letter_vectors[1,:], 1) * letter_vectors[0,:] * N ** 0.5
print ab, norm(ab)
print ba, norm(ba)
```

The norms of the bound vectors get noisy quite quickly. This noise is going to be a big source of our problems. However, if we can renormalize after taking the inverse, then we can get the right answers.



```python
inv_b1 = np.roll(letter_vectors[1,:] ** -1, 1)

ab_deref_prob = np.dot(letter_vectors, inv_b1 * ab / norm(inv_b1 * ab))
ba_deref_prob = np.dot(letter_vectors, inv_b1 * ba / norm(inv_b1 * ba))

figure(figsize=(10,4))
subplot(121)
bar(arange(len(ab_deref_prob)), ab_deref_prob)
ylim([0, 1])
subplot(122)
bar(arange(len(ba_deref_prob)), ba_deref_prob)
ylim([0, 1])
```

**Figure 2.11. Normalization enables multiplication in HRR:** Two hypervectors were formed, one by binding the sequence $ab$, the other by binding $ba$. These hypervectors were then dereferenced by the inverse of $r(b)$ in the first position. When the $ab$ hypervector is dereferenced (and the inverse result is normalized) then all of the dot products are small (because there is no hypervector that begins with b in the $ab$ hypervector). When $ba$ is dereferenced with $r(b)$, then the correct result, $a$ has a large dot product.

So, this is what we expect to happen, but this crucially depends on `norm(inv_b1 * ab)`. If we remove this norm and replace it by the expected normalization, then multiplication inverse begins to fall apart.

```python
inv_b1 = np.roll(letter_vectors[1,:] ** -1, 1)

ab_deref_prob = np.dot(letter_vectors, inv_b1 * ab / N ** 0.5)
ba_deref_prob = np.dot(letter_vectors, inv_b1 * ba / N ** 0.5)

figure(figsize=(10,4))
subplot(121)
bar(arange(len(ab_deref_prob)), ab_deref_prob)
#ylim([0, 1])
subplot(122)
bar(arange(len(ba_deref_prob)), ba_deref_prob)
#ylim([0, 1])
```

**Figure 2.12. Without normalization, multiplicative inverse explodes null values:** The same procedure was performed, but using the expected normalization value, rather than the true norm. When an inverse is multiplied by another hypervector, the extremely large values of the inverse can bleed through as hugh amounts of noise (compare left plots of 2.11 and 2.12). 

With the expected renormalization value ($1/\sqrt N$), using the inverse (`inv_b1`) to dereference $ab$ should return a lot of 0 values. But the noise is amplified because the norm of `inv_b1` is very noisy.

### 2.2 Complex HRR

Plate discusses also the use of hypervectors in the frequency domain and discusses the connection with this and Kanerva's binary splatter codes. 

In the complex domain, the hypervectors each have unit norm, but with arbitrary angle. The Kanerva binary vectors are simply a special case of the Plate vectors in the frequency domain -- where the angles are confined to only being $[0, \pi]$. These are also the special case where there is the self-inverse.

In the book, Plate discusses maintaining the structure of the hypervector throughout -- i.e. the Gaussian vectors stay Guassian, and the complex hypervectors stay unitary. This would also imply that the Kanerva vectors should stay $[-1, +1]$. However, when we build-up statistics into a large hypervector with memory, we do not constrain the values to be just $[-1, +1]$, but can be any integer. This may be "breaking the rules", but I think this is fine. 

When Plate stores only the phase of complex hypervectors, then there is an issue with $+$, because each element in the hypervector is always norm 1. This means that storing several elements get noisy much more quickly. If we allow both phase and magnitude in a complex hypervector, then we can do everything!

First, we are going to Plate's normal framework but add a new operation -- pconv, which is roll and multiply in the fourier domain. However, this will return a complex vector when binding to real valued vectors.

```python
def pconv(a, b, r=1):
    return ifft(np.roll(fft(a), r) * fft(b))
```

## Discussion

From these experiments we can see that Plate and Kanerva are extremeley related. Kanerva happens to be a special case of Plate, and the operations of Kanerva ($r$, $*$) apply to the Plate representation as well. But the inverse of multiplication is not stable in Plate's representation ($[\mathscr{N}(0,1/N)]^N$).

Most importantly, we see that Plate's binding operation does not contain the full flexibility of Kanerva's. This is because there is not a well defined permute operation that pairs with the binding operation. In the Plate framework, circular convolution is used as binding, but due to the commutative nature of permutation through circular convolution, there is a problem binding ordered vectors into a unique hypervector. 

Another factor is that in different frameworks vectors can be truly unitary (unit-norm) or only expected unitary (expected unit-norm). With the Plate framework, the hypervectors are unit-norm, but this gives only *expected unitary* in the fourier domain. The noise in the expectations can cause big impacts -- especially if the inverse of multiply is being used on Plate vectors, or if circular convolution is used repeatedly without renormalization (i.e. the small amount of noise is amplified by exponential growth).
